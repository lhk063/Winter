```python
import sys
import cv2


print('Hello, OpenCV', cv2.__version__) # 버전 출력

# img = cv2.imread('cat.bmp') # 원본 컬러 RGB

img = cv2.imread('cat.bmp', cv2.IMREAD_GRAYSCALE) # 그레이스케일 영상으로 일기

# 영상 파일 속성 그대로 읽기 Ex) PNG 파일에는 투명 파일이 있는데 이 특성으로 읽어주어야 4차원으로 반영이 되어서 읽을 수 있다.
# img = cv2.imread('cat.bmp', cv2.IMREAD_UNCHANGED) 


if img is None: # 예외 처리 코드를 작성하는게 좋다
    
    print('Image load failed!')
    sys.exit()


# 이미지 저장하기
cv2.imwrite('cat_gray.png', img)


# 새 창 띄우기
cv2.namedWindow('image') # Defalut = autosize

# cv2.namedWindow('image', cv2.WINDOW_NORMAL) # 마우스로 사이즈 조절이 가능하다.


cv2.imshow('image', img)

# key = cv2.waitKey() # keyboard 입력을 기다린다. 키보드를 누르기 전까지 영상이 켜져있다.
# print(key) # 아스키코드가 출력된다! ex) 27(ESC), 13(ENTER), 9(TAB)


# 특정 키를 누르기 전까지 켜져있다
while True:

    if cv2.waitKey() == 27: # 아스키 코드로 끄는법
        break 

# while True:

#     if cv2.waitKey() == ord('q'): # 특정 키
#         break 

# 창 닫기

# cv2.destroyWindow('image') # 지정한 창 닫기
cv2.destroyAllWindows() # 열려있는 모든 창을 닫기.
```

```python
import matplotlib.pyplot as plt
import cv2 as cv

# 컬러 영상 출력
imgBGR = cv.imread('cat.bmp')
imgRGB = cv.cvtColor(imgBGR, cv.COLOR_BGR2RGB)

plt.axis('off')
plt.imshow(imgRGB)
plt.show()
```

```python
# 그레이스케일 영상 출력
imgGray = cv.imread('cat.bmp', cv.IMREAD_GRAYSCALE)

plt.axis('off')
plt.imshow(imgGray, cmap = 'gray')
plt.show()
```

```python
# 두 개의 영상을 함께 출력
plt.subplot(121), plt.axis('off'), plt.imshow(imgRGB)
plt.subplot(122), plt.axis('off'), plt.imshow(imgGray, cmap = 'gray')
plt.show()
```
